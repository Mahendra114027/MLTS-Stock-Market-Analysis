## Exploratory Data Analysis of Stock Market Data

## Imports


```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from pylab import rcParams
rcParams['figure.figsize'] = 10, 6
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
import seaborn as sns
```

##Load Data

```python
df=pd.read_csv('TSLA.csv')
```

```python
df.head()
```

## Preprocess

## Converting the type of the Date Column

```python
df['Date']=pd.to_datetime(df["Date"])
```

###Column Names

```python
df = df.copy()
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head() 
```

## Exploratory Data Analysis

## Shape and Size

```python
print(f'Dataframe Shape: {df.shape}')
print(f'Dataframe Size: {df.size}') 
```

## Missing Values


```python
df.isnull().sum() 
```

## Columns skew


```python
df.skew(numeric_only=True) 
```

## Correlation


```python
correlations = df.corr()
correlations 
```

```python
sns.heatmap(df.corr())
plt.show()
```

## Seasonal Trends

```python
# Last n days visualized
n = 1000
result_data = seasonal_decompose(
    df.adj_close.tail(n),
    model='additive',
    period=30
)

fig = go.Figure()
fig = result_data.plot()
fig.set_size_inches(20, 19)
```

## Volume over Date

```python
df1=df.copy()
```

```python
df1.set_index('date', drop=True, append=False, inplace=True, verify_integrity=False)
```

```python
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Adj Close Prices')
plt.plot(df1['volume'])
plt.title('Tesla Corporation GROUP closing price')
plt.show()
```
